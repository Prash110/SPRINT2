```python
import mysql.connector
from mysql.connector import Error

try:
    # Database connection
    connection = mysql.connector.connect(
        host='localhost',       # Replace with your MySQL host
        user='root',            # Replace with your MySQL user
        password='9535', # Replace with your MySQL password
        database='demo'       # Replace with your database name
    )
    cursor = connection.cursor()
    print("Connected to the database successfully.")

    # Create the Flipkart table
    create_table_query = """
    CREATE TABLE Flipkart (
        Product_ID INT PRIMARY KEY,
        Product_Title VARCHAR(227),
        Product_Rating FLOAT,
        Category VARCHAR(50),
        Discounted_Price DECIMAL(10, 2),
        Actual_Price DECIMAL(10, 2),
        Discount_Percentage DECIMAL(5, 2)
    );
    """
    cursor.execute(create_table_query)
    print("Flipkart Table Created Successfully.")

except Error as e:
    print("Error:", e)

finally:
    if connection.is_connected():
        cursor.close()
        connection.close()
        print("Database connection closed.")
```

```python
import pandas as pd

# Load the CSV file
file_path = "C:\\abhi\\Flipkart 1.csv"
data = pd.read_csv(file_path)
data['Discount_Percentage'] = data['Discount_Percentage'].str.replace('%', '', regex=False).astype(float)
# Fill missing values with the mean for numeric columns
numeric_columns = ['Product_ID', 'Product_Rating', 'Discounted_Price', 'Actual_Price', 'Discount_Percentage']
for column in numeric_columns:
    data[column].fillna(data[column].mean(), inplace=True)

# Fill missing values with the mode for non-numeric columns
non_numeric_columns = ['Product_Title', 'Category']
for column in non_numeric_columns:
    data[column].fillna(data[column].mode()[0], inplace=True)
# Truncate Product_Title to a max length of 255 characters
data['Product_Title'] = data['Product_Title'].str.slice(0, 255)

# Check if all missing values are handled
print(data.isnull().sum())
```

```python
import mysql.connector
from mysql.connector import Error

try:
    # Database connection
    connection = mysql.connector.connect(
        host='localhost',
        user='root',
        password='9535',
        database='demo'
    )
    cursor = connection.cursor()
    print("Connected to the database successfully.")

    # Insert data into the Flipkart table
    insert_query = """
    INSERT INTO Flipkart (Product_ID, Product_Title, Product_Rating, Category, Discounted_Price, Actual_Price, Discount_Percentage)
    VALUES (%s, %s, %s, %s, %s, %s, %s)
    """
    
    # Loop through cleaned data
    for index, row in data.iterrows():
        cursor.execute(insert_query, (
            int(row['Product_ID']),  # Ensure type compatibility
            row['Product_Title'],
            float(row['Product_Rating']),
            row['Category'],
            float(row['Discounted_Price']),
            float(row['Actual_Price']),
            float(row['Discount_Percentage'])
        ))
        if index % 1000 == 0:  # Commit in batches of 1000
            connection.commit()
    
    connection.commit()  # Final commit
    print(f"{len(data)} rows inserted successfully into the Flipkart table.")

except Error as e:
    print("Error:", e)

finally:
    if connection.is_connected():
        cursor.close()
        connection.close()
        print("Database connection closed.")
```

```python
try:
    # Reconnect to the database
    connection = mysql.connector.connect(
        host='localhost',
        user='root',
        password='9535',
        database='demo'
    )
    cursor = connection.cursor()
    print("Connected to the database successfully.")

    # Fetch data from the Flipkart table
    select_query = "SELECT * FROM Flipkart;"
    cursor.execute(select_query)
    db_data = cursor.fetchall()  # Fetch all rows

    # Get column names from the database table
    column_names = [desc[0] for desc in cursor.description]

    # Create a pandas DataFrame
    flipkart_df = pd.DataFrame(db_data, columns=column_names)
    print("Data fetched successfully into a DataFrame.")
    print(flipkart_df.head())  # Preview the DataFrame

except Error as e:
    print("Error:", e)

finally:
    if connection.is_connected():
        cursor.close()
        connection.close()
        print("Database connection closed.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.histplot(flipkart_df['Product_Rating'], color='skyblue', bins=10)
plt.title('Distribution of Product Ratings')
plt.ylabel('Frequency')
plt.xlabel('Product Rating')
plt.show()
```

```python
category_discount_avg = flipkart_df.groupby('Category')['Discount_Percentage'].mean()
category_discount_avg.plot(kind='bar', figsize=(10,6), color='coral')
plt.title('Average Discount Percentage by Category')
plt.ylabel('Average Discount Percentage')
plt.xlabel('Category')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Product_Rating', y='Discount_Percentage', data=flipkart_df, hue='Category', palette='Set2', s=100)
plt.title('Product Rating vs Discount Percentage')
plt.xlabel('Product Rating')
plt.ylabel('Discount Percentage')
plt.show()
```

```python
# Correlation Matrix
corr = flipkart_df[['Product_Rating', 'Discounted_Price', 'Actual_Price', 'Discount_Percentage']].corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
category_count = flipkart_df['Category'].value_counts()
category_count.plot(kind='bar', figsize=(10,6), color='purple')
plt.title('Product Count by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
